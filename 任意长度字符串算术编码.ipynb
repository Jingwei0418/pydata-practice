```python
def cut_text(text,lenth): 
    textArr = re.findall('.{'+str(lenth)+'}', text) 
    textArr.append(text[(len(textArr)*lenth):]) 
    return textArr 
inf = 'cadacdbabaccbaacaccbaabadcdbabaccbaacaccbaababdcbabaccbaacabaccbaacaccbaaaccabaccbaacabaadababaccbaacaccbaadcbadcbabdcccbaaadbcdbcada'
infos = cut_text(inf,18)
```

```python
words = ['a', 'b', 'c','d']
freqs = [0.1, 0.4, 0.2, 0.3]
nc = len(words)
```

## 编码

```python
codes=[]
for info in infos:
    minv = 0
    maxv = 1
    for c in info:
        qujian = [i for i in range(nc+1)]
        qujian[0] = minv
        for j in range(1, nc+1):
            qujian[j] = qujian[j-1] + freqs[j-1]*(maxv-minv)
        print(qujian)
        new = words.index(c)
        minv = qujian[new]
        maxv = qujian[new+1]
    code = minv
    print('\n' + info + '的code: ' + str(code))
    codes.append(code)
```
